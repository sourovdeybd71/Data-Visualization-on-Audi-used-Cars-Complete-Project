```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("talk")
```

```python
df=pd.read_csv("audi.csv")
display(df.head())
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(y=df["mileage"],  showfliers=False)

plt.show()
```

```python
plt.figure(figsize=(4,4))
# adding a hue
sns.boxplot(y=df["mileage"],  x=df["fuelType"],showfliers=False)

plt.show()
```

# From a box-plot we can get:
min<br>
first quartile (25th percentile)
<br> median (50th percentile)
<br>third quartile (75th percentile)
<br> max
<br>range = max-min
<br>IQR (inter-quartile range) = third quartile - first quartile

#### separate the numerical columns

```python
def separate_num_cols(dataframe):
    #create an empty list to keep the numeric columns
    num_cols = []
    
    for col in dataframe.columns:
        if(dataframe[col].dtypes == 'int64') | (df[col].dtypes== 'float64'):
            num_cols.append(col)
            
    return num_cols

numeric_cols = separate_num_cols(df)
print(numeric_cols)
    
```

#### Box-plots of all numeric features

```python
plt.figure(figsize=(20,6))
for index in range(len(numeric_cols)):
    plt.subplot(2,3,index+1)
    sns.boxplot(x=df[numeric_cols[index]])
    
plt.suptitle("Boxplot of the numerical variables in the telecom dataset")
plt.tight_layout()
plt.show()
```
