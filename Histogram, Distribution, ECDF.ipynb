### Data Visualization: Histogram, Distribution, ECDF


```python
#library for data analysis
import pandas as pd
import numpy as np

#library for data visualization
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_context("talk")
```

```python
df = pd.read_csv("audi.csv")
display(df.head())
```

```python
df.describe()
```

```python
plt.figure(figsize=(20,12))

plt.subplot(2,2,1)
sns.histplot(data=df, x="mpg")
plt.title("Bydefault seaborn binwidth")

plt.subplot(2,2,2)
sns.histplot(data=df, x="mpg", binwidth=20)
plt.title("With setting Binwidth=20")

plt.subplot(2,2,3)
sns.histplot(data=df, x="mpg", binwidth=20)
plt.xticks(np.arange(20,200,20), rotation=90)
plt.title("Using arange method")

plt.subplot(2,2,4)
sns.histplot(data=df, x="mpg", binwidth=20, color="orange")
plt.xticks(np.arange(20,200,20))
plt.title("Using arange method")



plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,12))

plt.subplot(2,2,1)
sns.histplot(data=df, x="mpg", binwidth=20)
plt.xticks(np.arange(20,200,20))
plt.title("MPG")

plt.subplot(2,2,2)
sns.histplot(data=df, x="mileage", binwidth=50)
plt.xticks(np.arange(20,200,20))
plt.title("Mileage")

plt.subplot(2,2,3)
sns.histplot(data=df, x="year", binwidth=50, color="red")
plt.xticks(np.arange(20,200,20))
plt.title("Year")

plt.subplot(2,2,4)
sns.histplot(data=df, x="price", binwidth=50, color="orange")
plt.xticks(np.arange(20,200,20))
plt.title("Price")


plt.tight_layout()
plt.show()
```

From above visualization we can see without 'mpg' column <br/> others are not fit by histogram.

```python
plt.figure(figsize=(20,12))

plt.subplot(2,2,1)
sns.histplot(data=df, x="mpg")
plt.title("Bydefault seaborn binwidth")

plt.subplot(2,2,2)
sns.histplot(data=df, x="mpg", binwidth=1, cumulative=True)
plt.title("With setting Binwidth=20")

plt.subplot(2,2,3)
sns.histplot(data=df, x="mpg", binwidth=1)
plt.grid()
plt.title("Using arange method")

plt.subplot(2,2,4)
sns.histplot(data=df, x="mpg", binwidth=1, cumulative=True, stat="probability", hue="fuelType")
plt.xticks(np.arange(20,200,20))
plt.title("Using arange method")


plt.tight_layout()
plt.show()
```
