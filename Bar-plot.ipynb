```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('audi.csv')
display(df)
```

```python
fuelType_count = df['fuelType'].value_counts(normalize=True)
#display(fuelType_count)

fuelType_count=pd.DataFrame(fuelType_count)
# display(fuelType_count)

fuelType_count=fuelType_count.reset_index()
# display(fuelType_count)

fuelType_count=fuelType_count.rename(columns={"index":"fuelType", "fuelType":"% of cars"})
# display(fuelType_count)

fuelType_count['% of cars']=np.round(fuelType_count["% of cars"]*100,2)
display(fuelType_count)
```

```python
plt.figure(figsize=(5,6))

sns.barplot(x='fuelType', y='% of cars', data= fuelType_count, color="blue")
plt.title("Percent of AUDI CARs present for each FUELTYPE ")

plt.xlabel("Audi car fueltypes")
plt.ylabel("percentage of cars")

plt.yticks(np.arange(0,101,20))
plt.grid()
plt.show()
```

```python
transmission_count = df['transmission'].value_counts(normalize=True)
#display(transmission_count)

transmission_count=pd.DataFrame(transmission_count)
# display(transmission_count)

transmission_count=transmission_count.reset_index()
# display(transmission_count)

transmission_count=transmission_count.rename(columns={"index":"transmission", "transmission":"% of cars"})
# display(transmission_count)

transmission_count['% of cars']=np.round(transmission_count["% of cars"]*100,2)
display(transmission_count)
```

```python
plt.figure(figsize=(5,6))

sns.barplot(x='transmission', y='% of cars', data= transmission_count, color="green")
plt.title("Percent of AUDI CARs present for each Transmission ")

plt.xlabel("Audi car Transmissions")
plt.ylabel("percentage of cars")

plt.yticks(np.arange(0,101,20))
plt.grid()
plt.show()
```

```python
model_count = df['model'].value_counts(normalize=True)
#display(model_count)

model_count=pd.DataFrame(model_count)
# display(model_count)

model_count=model_count.reset_index()
# display(model_count)

model_count=model_count.rename(columns={"index":"model", "model":"% of cars"})
# display(model_count)

model_count['% of cars']=np.round(model_count["% of cars"]*100,2)
display(model_count)
```

```python
plt.figure(figsize=(5,6))

sns.barplot(x='model', y='% of cars', data= model_count, color="red")
plt.title("Percent of AUDI CARs present for each MODELs ")

plt.xlabel("Audi car MODELs")
plt.ylabel("percentage of cars")

plt.yticks(np.arange(0,31,10))
plt.xticks(rotation=90)

plt.grid()
plt.show()
```

```python
year_count = df['year'].value_counts(normalize=True)
#display(year_count)

year_count=pd.DataFrame(year_count)
# display(year_count)

year_count=year_count.reset_index()
# display(year_count)

year_count=year_count.rename(columns={"index":"year", "year":"% of cars"})
# display(year_count)

year_count['% of cars']=np.round(year_count["% of cars"]*100,2)
display(year_count)
```

```python
plt.figure(figsize=(5,6))

sns.barplot(x='year', y='% of cars', data= year_count, color="orange")
plt.title("Percent of AUDI CARs present for each MODELs ")

plt.xlabel("Audi car YEARs")
plt.ylabel("percentage of cars")

plt.yticks(np.arange(0,31,10))
plt.xticks(rotation=90)

plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16,12))

plt.subplot(2,2,1)
sns.barplot(x='fuelType', y='% of cars', data= fuelType_count, color="blue")
plt.title("Percent of AUDI CARs present for each FUELTYPE ")

plt.xlabel("Audi car fueltypes")
plt.ylabel("percentage of cars")
plt.yticks(np.arange(0,101,20))


plt.subplot(2,2,2)
sns.barplot(x='transmission', y='% of cars', data= transmission_count, color="green")
plt.title("Percent of AUDI CARs present for each Transmission ")

plt.xlabel("Audi car Transmissions")
plt.ylabel("percentage of cars")
plt.yticks(np.arange(0,101,20))


plt.subplot(2,2,3)
sns.barplot(x='model', y='% of cars', data= model_count, color="red")
plt.title("Percent of AUDI CARs present for each MODELs ")

plt.xlabel("Audi car MODELs")
plt.ylabel("percentage of cars")

plt.yticks(np.arange(0,31,10))
plt.xticks(rotation=90)


plt.subplot(2,2,4)
sns.barplot(x='year', y='% of cars', data= year_count, color="orange")
plt.title("Percent of AUDI CARs present for each MODELs ")

plt.xlabel("Audi car YEARs")
plt.ylabel("percentage of cars")

plt.yticks(np.arange(0,31,10))
plt.xticks(rotation=90)

plt.suptitle("Categorical columns in the BMW used CARS")
plt.tight_layout()
plt.show()
```

```python
def unique_percentage_count(data, col):
    col_count = data[col].value_counts(normalize=True)
    col_count=pd.DataFrame(col_count)
    col_count=col_count.reset_index()
    col_count=col_count.rename(columns={"index":col, col:"% of cars"})
    col_count['% of cars']=np.round(col_count["% of cars"]*100,2)
    return col_count

year_count= unique_percentage_count(df, "year")
display(year_count)

fuelType_count= unique_percentage_count(df, "fuelType")
display(fuelType_count)

model_count= unique_percentage_count(df, "model")
display(model_count)

transmission_count= unique_percentage_count(df, "transmission")
display(transmission_count)
```

```python
sns.set_context("talk")
plt.figure(figsize=(15,10))

def barplot(x,y,data,color, rotation):
    sns.barplot(x=x, y=y, data= data, color=color)
    plt.title("Percent of AUDI CARs present for each "+x)

    plt.xlabel("Audi car MODELs")
    plt.ylabel("percentage of cars")

    plt.yticks(np.arange(0,101,10))
    plt.xticks(rotation=rotation)
    
plt.subplot(2,2,1)
barplot("model", "% of cars", model_count, "blue", 90)

plt.subplot(2,2,2)
barplot("year", "% of cars", year_count, "red", 90)

plt.subplot(2,2,3)
barplot("transmission", "% of cars", transmission_count, "orange", 0)

plt.subplot(2,2,4)
barplot("fuelType", "% of cars", fuelType_count, "green", 0)

plt.suptitle("Categorical columns in the BMW used CARS")
plt.tight_layout()
plt.show()
```
